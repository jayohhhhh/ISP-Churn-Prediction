# Modeling

### Feature Selection
All features will be tested in these models including the "is_contract" that was not included in the original dataset as they all show positive or negative relevant correlation to "churn".

```python
xvars = net[['is_tv_subscriber', 'subscription_age', 'bill_avg', 'remaining_contract', 'service_failure_count', 'download_avg', 'upload_avg', 'is_contract']]
```

```python
X = xvars
y = net['churn']
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=2347)
```

### Models

```python
rf = RandomForestClassifier()
lr = LogisticRegression(max_iter=10000)
tree = DecisionTreeClassifier(max_depth=4, min_samples_split=5)
ada = AdaBoostClassifier()
bag = BaggingClassifier()
gb = GradientBoostingClassifier()
xg = xgb
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf.score(X_train, y_train), rf.score(X_test, y_test)
```

```python
lr = LogisticRegression(max_iter=10000)
lr.fit(X_train, y_train)
lr.score(X_train, y_train), lr.score(X_test, y_test)
```

```python
tree = DecisionTreeClassifier(max_depth=4, min_samples_split=5)
tree.fit(X, y)
tree.score(X_train, y_train), tree.score(X_test, y_test)
```

```python
bag = BaggingClassifier()
bag.fit(X_train, y_train)
bag.score(X_train, y_train), bag.score(X_test, y_test)
```

```python
gb = GradientBoostingClassifier(random_state=234)
gb.fit(X_train, y_train)
gb.score(X_train, y_train), gb.score(X_test, y_test)
```

```python
xg = xgb.XGBClassifier()
xg.fit(X_train, y_train)
xg.score(X_train, y_train), xg.score(X_test, y_test)
```

```python
sc = StandardScaler()
Z_train = sc.fit_transform(X_train)
Z_test = sc.transform(X_test)
```

```python
knn = KNeighborsClassifier()
knn.fit(Z_train, y_train)
knn.score(Z_train, y_train), knn.score(Z_test, y_test)
```

### Loop Through Models

```python
models = (rf, lr, tree, ada, bag, gb)
for model in models:
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    print(f'{model} Train score: {model.score(X_train, y_train)}')
    print(f'{model} Test mse: {model.score(X_test, y_test)}')
    print(f'{model} F1 Train score: {f1_score(y_train, y_train_pred)}')
    print(f'{model} F1 Test score: {f1_score(y_test, y_test_pred)}')
    print(f'{model} Train mse: {np.sqrt(mean_squared_error(y_train, y_train_pred))}')
    print(f'{model} Test mse: {np.sqrt(mean_squared_error(y_test, y_test_pred))}')
```

```python
importances = gb.feature_importances_
feature_names = np.arange(X.shape[1])
most_important_idx = np.argmax(importances)
most_important_feature = feature_names[most_important_idx]
print("Most important feature:", most_important_feature)
#source cite: ChatGPT for most important feature code
```

```python
# most important features
pd.DataFrame({
    'feature': X.columns,
    'importances': gb.feature_importances_
}).sort_values('importances', ascending=False)
```

## Model of Choice: GradientBoost w/ GridSearch

- GradientBoost model had slightly higher scores than all other models while proving not overfit despite its tendencies. Also yielded the best accuracy improvements when gridsearching.
- Model performance increased by about 1% by increasing n_estimators from 20 to 150, max_features to log2, and max_depth to 7.
- Adding the "is_contract" feature increased model performance by about 1%.

```python
params = {
    'learning_rate': [0.1],
    'n_estimators': [150],
    'max_depth': [7],
    'max_features': ['log2']
}

gs = GridSearchCV(gb, params, cv=5, n_jobs=6, verbose=0)
gs.fit(X_train, y_train)
```

```python
gs.score(X_train, y_train), gs.score(X_test, y_test)
```

```python
gs.best_params_
```

```python
preds = gs.predict(X_test)
```

Confusion matrix is on the test set which is 25% of our data(21,568 rows) and it predicted 11,000 True Positives(51%). These are customers for which we could offer extended contracts or promotions to retain them.

We have 530 False Positives(0.02%) where we could lose money since the promotion or extension of contract offered to these customers will be an extra cost. 

740 are True Negatives(0.03%) which is a loss of good customers.

## Conclusion and Future Steps

Customers that had no internet service contract churned over 90% of the time. Customers with no TV subscription, over 85% of the time. Third most important factor was Movie Package Subscriptions. The first steps that should be taken to reduce customers from churning is for the business to focus on providing as much value in these 3 products to the customers as possible as the data shows, the longterm customers that can be gained can easily outweigh the cost of these product investments.
